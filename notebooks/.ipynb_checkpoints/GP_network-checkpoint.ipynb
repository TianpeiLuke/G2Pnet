# Gaussian Processs in network
## Motivations
In this project, we consider latent representations of nodes in a network using a probabilistic model. We assume that each node attribute vector $\mathbf{x}_{v}$ is generated  from a probabilistic latent variable model  $P(\mathbf{x}| \mathbf{z}, \mathcal{E})$, where $\mathcal{E}$ defines a network topology $\mathcal{G} = (\mathcal{V}, \mathcal{E})$ and $\mathbf{x}_{v}$ is the attribute for node $v\in \mathcal{V}$. 

Our goal is to predict an unobserved node attribute vector $\mathbf{x}_{u}$ given all the attributes from the other observed nodes $\mathbf{x}_{\mathcal{V}}$ and the graph topology $\mathcal{E}$.

Consider the $k$-dimensional eigen-subspace $\mathbf{U} \in G(|\mathcal{V}|, k) \equiv \mathcal{O}(|\mathcal{V}|)/\mathcal{O}(k) \times \mathcal{O}(|\mathcal{V}|-k)$ of the (normalized) Laplacian $\mathbf{L} \equiv  \mathbf{I} - \mathbf{A}_{W}^{-1/2}\mathbf{A}\mathbf{A}_{W}^{-1/2} $, where each row of $\mathbf{U}$, $\mathbf{u}_{n}^{\tau} \in \mathbb{R}^{k}$ provides a coordinate of each node in the graph under a set of spectral axes. Then a summmary of the graph topology can be obtained by projecting the spectral layout of the node into a one-dimensional subspace defined by $\mathbf{s}_{v}$, i.e., 
\begin{align}
\mathbf{z}_{v} =  \mathbf{U}^{T}\mathbf{s}_{v}. 
\end{align}

Finally, $\mathbf{z}_{v}$ is seen as the input features to generate the node attributes $\mathbf{x}_{v}$. 


A common question is to ask how much information regarding the topology of the network is incorporated. It is especially important if the node is sub-sampled from the graph.

The above figure shows a barbell graph which consists of two complete graphs connected by a path of length $0$ or $1$. The two-dimensional spectral coordinate of the graph is shown on the right hand size. The graph is generated by Random Walks on Graphs.

generate the eigenspace of normalized Laplacian

```python
#generate the eigenspace of normalized Laplacian
k = 3
nnodes = nx.adjacency_matrix(G0).shape[0]

ncv=max(2*k+1,int(np.sqrt(nnodes)))

eigenvalues,eigenvectors= sp.sparse.linalg.eigsh(nx.normalized_laplacian_matrix(G0), k=k, \
                                                which='SM', return_eigenvectors = True, ncv=ncv)
index=np.argsort(eigenvalues)[1:k] # 0 index is zero eigenvalue    
pos0_mat = np.real(eigenvectors[:,index])
```

### Choose $\mathbf{s}$ as random projection according to isotropic Normal distribution

```python
N = pos0_mat.shape[0]
d = pos0_mat.shape[1]

T = 10*N
#z = np.zeros([T, 2])
s = np.random.randn(N,T)
z = np.dot(pos0_mat.T, s).T 
```

```python
fig = plt.figure(3)
plt.scatter(z[:,0], z[:,1])
plt.title('Scatter plot of projected vector z', fontsize = 15)
plt.ylabel('axis 1', fontsize=15)
plt.xlabel('axis 0', fontsize=15)
plt.grid()
plt.show()
```

```python
ax = sns.kdeplot(z[:,0], z[:,1])
```

We see that after projections by $T$ i.i.d. Gaussian random directions, we obtained a set of $2$-d points 
\begin{align} 
    \mathbf{Z} = [\mathbf{z}_{1}, \ldots, \mathbf{z}_{T}] &= \mathbf{U}^{T}[\mathbf{s}_{1}, \ldots, \mathbf{s}_{T}]
\end{align}

Given $\mathbf{s}_{v} \sim \mathcal{N}(\mathbf{0}, \mathbf{I})$, we know that since $\mathbf{U}^{T}\mathbf{U} = \mathbf{I}_{k}$, 
\begin{align}
\mathbf{z}_{v} & \sim \mathcal{N}(\mathbf{0}, \mathbf{I}_{k})
\end{align}

This looks disappointing, since the clustering information in __rows__ of $\mathbf{U}$ does not show up. That is, the non-informatic _isotropic_ prior on $\mathbf{s}$ cannot reveal the true information about the clustering. 

So __what can we do ?__ 

Note that the isotropic prior $\mathbf{s}_{v} \sim \mathcal{N}(\mathbf{0}, \mathbf{I})$ has the maximal entropy. The opposite extreme case is when the entropy is smallest, i.e., ask $\mathbf{s} \in \{0,1\}$. 

### Choose $\mathbf{s}$ as coordinate projection

```python
N = pos0_mat.shape[0]
d = pos0_mat.shape[1]
s0 = np.zeros([N,d])
s0[np.arange(0,int(N/2)),0] = np.ones([int(N/2),])
s0[int(N/2)+np.arange(0,int(N/2)),1] = np.ones([int(N/2),])
s = np.concatenate((np.tile(s0[:,0].T,(int(N/2),1)).T, np.tile(s0[:,1].T,(int(N/2),1)).T), axis=1)

z0 = np.dot(pos0_mat.T, s).T 
z0.shape
```

In $\mathbf{s}^{0}$, we have first $N/2$ columns are $[1,0]^{T}$ and the last $N/2$ columns are $[0,1]^{T}$. After the projection, we see that 

```python
fig = plt.figure(4)
plt.scatter(z0[:,0], z0[:,1])
plt.title('Scatter plot of projected vector z', fontsize = 15)
plt.ylabel('axis 1', fontsize=15)
plt.xlabel('axis 0', fontsize=15)
plt.grid()
plt.show()
#s0
```

In this plot, we see that the input feature $z^0$ only takes two values, each corresponding to a cluster. This is because all information about the cluster is encoded in first coordinate of $\mathbf{U}$, which can be extracted for $\mathbf{s}^{0} = [1,0]$. 

### Choose  __s__  as random projection from non-isotropic Gaussian Process

```python
from sklearn.metrics.pairwise import rbf_kernel, pairwise_kernels
N = pos0_mat.shape[0]
d = pos0_mat.shape[1]

T = 10*N
latent_index = np.arange(N) - N/2
#Kernel = rbf_kernel(X=latent_pts, gamma=100)

Kernel = pairwise_kernels(X=latent_index.reshape(-1, 1), metric = 'sigmoid', n_jobs = 8, gamma= 50/N) 
```

Given $\mathbf{s}_{v} \sim \mathcal{N}(\mathbf{0}, \mathbf{\Sigma})$, we know that since $\mathbf{U}^{T}\mathbf{U} = \mathbf{I}_{k}$, 
\begin{align}
\mathbf{z}_{v} & \sim \mathcal{N}(\mathbf{0}, \mathbf{U}^{T}\mathbf{\Sigma}\mathbf{U})
\end{align}

## Models
In particular, the node representation is given by 
\begin{align*}
 \mathbf{x}_{v} &= g(\mathbf{U}^{T}\mathbf{s}_{v}) + \mathbf{n}
\end{align*}

 That is, given all node attributes $\mathbf{x}_{\mathcal{N}}$ in $\mathcal{G}$ and the graph topology $\mathcal{E}$, the joint distribution for test sample $\mathbf{x}_{v}$ and  $\mathbf{x}_{\mathcal{N}}$ follows 
\begin{align*}
[\mathbf{x}_{v}, \mathbf{x}_{\mathcal{N}}] &\sim \mathcal{N}(\mathbf{0}, \mathbf{K}_{\mathcal{E}}(\mathbf{Z})),
\end{align*} where 
\begin{align*}
 \mathbf{K} &= \left[K(\mathbf{z}_{n}, \mathbf{z}_{m})\right]
\end{align*}

Here, $\mathbf{Z} = [\mathbf{z}_{1}, \ldots, \mathbf{z}_{|\mathcal{V}|}]$ is the latent representation of nodes in network $\mathcal{G}$.
