## The manifold (graph) regularization 
It is also well-known that in the manifold-regularized semi-supervised learning model
\begin{align}
\hat{f} &= \arg\min\limits_{f\in \mathcal{H}_{K}} \frac{1}{n}\sum_{i=1}^{n}V(f(\mathbf{x}_{i}), y_{i}) + \gamma_{A}\|f\|_{K}^{2} + \gamma_{I}\|f\|_{\mathcal{M}}^{2}, 
\end{align}
where $$
\|f\|_{\mathcal{M}}\equiv \int_{\mathcal{M}}\|\nabla_{\mathcal{M}} f\|^{2}dP_{X}(x) = \int_{\mathcal{M}}f(x)\Delta_{\mathcal{M}} f(x) \;dP_{X}(x)
$$ defines an induced norm from the manifold $\mathcal{M}$ (given in Laplacian of K-nearest-neighbor Graph). 

The Representer theorem gives that
\begin{align}
\hat{f} &= \hat{f}_{K} + \hat{f}_{\mathcal{M}}
\end{align}
where $ \hat{f}_{K} = \sum \alpha_{i}K(\mathbf{x}_{i}, \cdot) = \sum_{k}\lambda_{k}<f, \phi_{k}>$ and 
\begin{align}
\hat{f}_{\mathcal{M}} &= \int_{\mathcal{M}}\alpha(z)K(x,z)dP_{X}(z) = \sum_{s}\mu_{s}<f, \hat{\phi}_{s}>
\end{align}
 $\{\phi_{k}\}$ is the set of eigen-basis of Hilbert space and $\{\hat{\phi}_{s}\}$ is the set of eigen-basis of Laplacian space

In kernel method, it compute a modified kernel
\begin{align}
\mathbf{K} &\leftarrow \mathbf{J}\mathbf{K}\left( 2\gamma_{A}\mathbf{I} + 2\frac{\gamma_{I}}{(l+u)^{2}} \mathbf{L}\mathbf{K}  \right)^{-1}\mathbf{J}^{T}
\end{align} where $\mathbf{J} = [\mathbf{I}, \mathbf{0}] \in \mathbb{R}^{l \times (l+u)}$ is the selection matrix.

## Learning node representation via Gaussian Process on network

We can define the Gaussian process on network domain using eigen-spaces. Let the Laplacian of network $\mathbf{L} \equiv  \mathbf{A}_{W} - \mathbf{A}$, where the adjacency matrix $\mathbf{A}$ is fully sampled. We assume that the nodes are sub-sampled so that their node attributes are missing. Our __goal__ is to predict their node attributes, given the graph topology $\mathcal{G} = (\mathcal{V}, \mathcal{E})$. 

We verified that features generated by Gaussian process using the eigenmatrix $\mathbf{U}$ of graph Laplacian $\mathbf{L}$ as input encode the cluster information. 

Now we reverse the direction. Assume that we have clustered graph dataset.

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import seaborn as sns
import pandas as pd
%matplotlib inline 

plt.close('all')
Ifsavefig = True
Ifsavedata = False
```

```python
size = 40
color_map = {0:'r', 1:'c', 2:'y'}
#center = np.zeros(2)
#center[1] = 1
center = None

n = 40
m = 0
G0= nx.barbell_graph(n,m) 
#generate the bar-bell graph, which consists of two fully connected component and a path connecting them 

label0 = np.zeros([2*n,1])
label0[n:2*n] = np.ones([n,1])
# labeling of the cluster
for node in G0.nodes():
    G0.node[node]['category'] = int(label0[node,0])
```

The Bar-Bell graph consists of two fully connected graphs with size $n$ and a path of length $m$. It guarantees that there are only two clusters and they are far from each other, which is ideal case. 

Now generate the node attribute vector for each node depending on their location on the graph. Let $\mathbf{x}^{v} = [x_{t}^{v}: t=1,\ldots,p]$, where $x_{t}^{v} = G(t; s_{v})$, and $s_v = p_x^v + j\,p_{y}^{v}$. Here $(p_x^v,  p_y^v)$ is the location of node in graph representation. 

Let  
\begin{align}
G(t; s_{v})&= \mathcal{Re}\{\exp(-s_{v}t)\}\\
&= \mathcal{Re}\{\exp(-p_x^v\; t)\left[\cos(p_y^v t) - j\sin(p_y^v t)\right]\}\\
&= \exp(-p_x^v\; t)\cos(p_y^v t),\quad t=1,\ldots, p
\end{align} and $x_{t}^{v} = G(t; s_{v}) + n^{v}(t)$
or
\begin{align}
\mathbf{G}(t; s_{v})&=\exp(-s_{v}t)\\
&= \exp(-p_x^v\; t)\left[\cos(p_y^v t) - j\sin(p_y^v t)\right]\\
&= \left[\begin{array}{c}
\exp(-p_x^v\; t)\cos(p_y^v t)\\
-\exp(-p_x^v\; t)\sin(p_y^v t)
\end{array}\right]
,\quad t=1,\ldots, p/2
\end{align} and $\mathbf{x}_{t}^{v} = \mathbf{G}(t; s_{v}) + \mathbf{n}^{v}(t)$

In other word, we project the graph with spectral layout onto the complex plane (Laplace domain) in which each point in Laplace domain corresponds to a continuous wave function with the maginitude and the frequency associated with the spectral position of that node. This wave function is our node feature.  

In general, the model for the node feature is 
\begin{align}
\mathbf{x}^{v} &= f(\mathbf{u}^{v}) + \mathbf{n}^{v}
\end{align} where $\mathbf{u}^{v}$ is the $v$-th row of eigenmatrix $\mathbf{U}$ of Laplacian.

```python
#generate the eigenspace of normalized Laplacian
k = 3
nnodes = nx.adjacency_matrix(G0).shape[0]

ncv=max(2*k+1,int(np.sqrt(nnodes)))

eigenvalues,eigenvectors= sp.sparse.linalg.eigsh(nx.laplacian_matrix(G0).asfptype(), k=k, \
                                                which='SM', return_eigenvectors = True, ncv=ncv)
index=np.argsort(eigenvalues)[1:k] # 0 index is zero eigenvalue    
pos_spectral_mat = np.real(eigenvectors[:,index]) #pos_spectral_mat is the spectral location of the nodes
pos_spectral_df = pd.DataFrame(data=pos_spectral_mat, columns=['PX','PY'], dtype=float)
pos_spectral_df['LABEL'] = label0
```

##  Model and Inference
After generated the node features according to their graph location, we need to build models. I use the Gaussian process. Let $f \in \mathcal{H}_{K}$, where $\mathcal{H}_{K}$ is the Repreducing Kernel Hilbert Space (RKHS) with kernel $K$. 

The joint distribution of node attributes
\begin{align}
[\mathbf{x}_{1}, \ldots, \mathbf{x}_{N}, \mathbf{x}_{v}]&= [f(\mathbf{u}_{1}), \ldots, f(\mathbf{u}_{N}), f(\mathbf{u}_{v})]\\
&\sim \mathcal{N}(\mathbf{0}, \mathbf{K})
\end{align} where 
\begin{align}
\left[
\begin{array}{cc}
\mathbf{K}_{-v,-v} & \mathbf{K}_{-v,v}\\
 \mathbf{K}_{v,-v} &  \mathbf{K}_{v,v}
\end{array}
\right]
\end{align} and $ \mathbf{K}_{n,m} = K(\mathbf{u}^{n}, \mathbf{u}^{m})$ for $m,n \in \mathcal{V}$.

```python
from sklearn.cross_validation import train_test_split
from sklearn.metrics.pairwise import rbf_kernel, pairwise_kernels
import GPy
```

```python
U = pos_spectral_df
U.drop(['LABEL'], axis=1, inplace= True)

Y = data_df
Y.drop(['LABEL'], axis=1, inplace= True)
```

Split data into training and testing

```python
U_train,  U_test, Y_train, Y_test = train_test_split(U, Y , test_size = 0.5, random_state = 45 )
U_train.shape
```

Call GPy package to optimize the kernel parameter.

```python
rbf_kern = GPy.kern.RBF(2)

model = GPy.models.GPRegression(X= U_train.values, Y = Y_train.values, kernel = rbf_kern)
model.Gaussian_noise.variance.constrain_bounded(1e-5, 0.1)
max_iters = 50
model.optimize(max_iters= max_iters, messages=True)
```

```python
print("Result: ")
print(model)
```

```python
mean_fun, var_fun = model.predict(U_test.values)
```

```python
sorted_index = np.argsort(U_test.index) #sort node index for plot 
error = 2*np.sqrt(var_fun[sorted_index]) # 95% confidence interval with 2-standard deviation

Y_show = Y_test.iloc[sorted_index,:].values
temp_min = mean_fun[sorted_index,:] - error
temp_max = mean_fun[sorted_index,:] + error

fig = plt.figure(4)
fig.set_size_inches(12.0,8.0)
plt.style.use('seaborn-white')

ax21 = plt.subplot(211)
h1 = plt.plot(U_test.index[sorted_index], Y_show[:,0], 'b')
h2 = plt.plot(U_test.index[sorted_index], mean_fun[sorted_index,0], 'r')
plt.fill_between(U_test.index[sorted_index], temp_min[:,0], temp_max[:,0], facecolor='#D3D3D3', alpha=0.5)
plt.plot((n-1)*np.ones([nnodes,]), \
         np.linspace(start=np.min(data_df[column_name[0]])-0.5, stop= np.max(data_df[column_name[0]])+0.5, num=nnodes),\
         'r', linewidth=2.0
        )
#plt.xlabel('node index', fontsize = 15)
plt.ylabel('node feature 1', fontsize = 15)
ax21.set_ylim([np.min(data_df[column_name[0]])-0.5, np.max(data_df[column_name[0]])+0.5])
ax21.legend(['true', 'predict'], fontsize = 15, loc='upper right', frameon=True)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()


ax22 = plt.subplot(212, sharex = ax21)
h1 = plt.plot(U_test.index[sorted_index], Y_show[:,1], 'b')
h2 = plt.plot(U_test.index[sorted_index], mean_fun[sorted_index,1], 'r')
plt.fill_between(U_test.index[sorted_index], temp_min[:,1], temp_max[:,1], facecolor='#D3D3D3')
plt.xlabel('node index', fontsize = 15)
plt.ylabel('node feature 2', fontsize = 15)
plt.plot((n-1)*np.ones([nnodes,]), \
         np.linspace(start=np.min(data_df[column_name[1]])-0.5, stop= np.max(data_df[column_name[1]])+0.5, num=nnodes),\
         'r', linewidth=2.0
        )
ax22.set_ylim([np.min(data_df[column_name[1]])-0.5, np.max(data_df[column_name[1]])+0.5])
ax22.legend(['true', 'predict'], fontsize = 15, loc='upper right', frameon=True)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid()


plt.show()

if Ifsavefig:
    fig.savefig('../figure/node_feature_GPpredict_03_23_16.eps')
```
