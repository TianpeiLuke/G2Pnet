# Compute the gradient of negative log-likelihood

The loss function is given as negative log-likelihood of Gaussian process

\begin{align}
\mathcal{L}(\beta, \sigma, \mathbf{U}) &= \frac{N}{2}\log(2\pi)+\frac{1}{2}\log\det(\mathbf{K}_{\beta, \sigma, \mathbf{U}})+\frac{1}{2}\text{tr}\left(\mathbf{K}_{\beta, \sigma, \mathbf{U}}^{−1}\mathbf{X}\mathbf{X}^T/p\right)
\end{align}
The partial differential of $\mathcal{L}$ with respect to $\mathbf{U}$
\begin{align}
d_{U}\mathcal{L} &= \frac{1}{2}\left[\text{tr}\left( \mathbf{T} \frac{\partial \mathbf{K}_{\beta, \sigma, \mathbf{U}}}{\partial \mathbf{U}_{m,j}} \right)\right]_{m,j},\\
\text{where }\mathbf{T} &= 2\mathbf{T}_{0} - \text{diag}\left(\mathbf{T}_{0}\right),\\
\mathbf{T}_{0} &= \left(\mathbf{I} - \mathbf{K}_{\beta, \sigma, \mathbf{U}}^{-1}\mathbf{X}\mathbf{X}^T/p \right)\mathbf{K}_{\beta, \sigma, \mathbf{U}}^{-1}, \\
\mathbf{K}_{r,s} &= K_{\sigma}\left(\mathbf{U}_{r,:} , \mathbf{U}_{s,:}\right) + \beta \delta_{r,s}
\end{align}

### Use einsum to compute the derivates according to the chain rule

```python
n = 100
k = 2

x = np.linspace(0, np.sqrt(n), np.sqrt(n))
y = np.linspace(0, np.sqrt(n), np.sqrt(n))
xv, yv = np.meshgrid(x, y)
Z = np.zeros([n,k])
Z[:,0] = xv[:].reshape([xv.size, ])
Z[:,1] = yv[:].reshape([yv.size, ])

p = 4

F = np.zeros([n,p])
F[:,0] = np.real(np.exp(0.2*Z[:,0]))
F[:,1] = np.real(np.exp(0.2*Z[:,0] + 1.j*Z[:,1]))
F[:,2] = np.real(np.exp(0.2*Z[:,0] + 2.j*Z[:,1]))
F[:,3] = np.real(np.exp(0.2*Z[:,0] + 3.j*Z[:,1]))
X = F + np.random.randn(n,p)
#Z, W = gutil.linalg.pca(X, k)
```

```python
rbf_kern = GPy.kern.RBF(2)+ GPy.kern.White(2)
kernel = rbf_kern

model = GPy.models.GPRegression(X= Z, Y = X, kernel = kernel)
model.Gaussian_noise.variance.constrain_bounded(1e-5, 2)
max_iters = 110
model.optimize(max_iters= max_iters, messages=True)
```

```python
print(model)
```

```python
lengthscale = model['sum.rbf.lengthscale'].values
variance = model['sum.rbf.variance'].values
kern = model.kern
Kernel = kern.K(Z,Z)
```

The partial differential of $\mathcal{L}$ with respect to $\mathbf{K}$
\begin{align}
\mathbf{T} = \frac{dL}{d\mathbf{K}} &= 2\mathbf{T}_{0} - \text{diag}\left(\mathbf{T}_{0}\right),\\
\text{where }\mathbf{T}_{0} &= \frac{1}{2}\left(\mathbf{I} - \mathbf{K}_{\beta, \sigma, \mathbf{U}}^{-1}\mathbf{X}\mathbf{X}^T/p \right)\mathbf{K}_{\beta, \sigma, \mathbf{U}}^{-1}, \\
\mathbf{K}_{r,s} &= K_{\sigma}\left(\mathbf{U}_{r,:} , \mathbf{U}_{s,:}\right) + \beta \delta_{r,s}
\end{align}

```python
S = gutil.linalg.tdot(X) 
KiX, _ = gutil.linalg.dpotrs(np.asfortranarray(Kernel), np.asfortranarray(S), lower=1)
# KiS = K^{-1}*X*X
KiXXiK, _ = gutil.linalg.dpotrs(np.asfortranarray(Kernel), np.asfortranarray(KiX.T), lower=1)
Ki,_ = gutil.linalg.dpotri(np.asfortranarray(Kernel), lower=1)
dL_dK_0 = 0.5*(p*Ki - KiXXiK)
dL_dK =  2*dL_dK_0 - np.diag(dL_dK_0)
dL_dK
```

```python
fig= plt.figure(2)
fig.set_size_inches(12.0, 12.0)
sns.set(style="white")

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

ax = plt.axes()
# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(dL_dK, cmap=cmap, vmax=np.max(np.max(dL_dK)), #.3,
            square=True, xticklabels=10, yticklabels=10,
            linewidths=.8, cbar_kws={"shrink": .5}, ax=ax)
#ax.set_title('sigmoid kernel', fontsize = 20)
plt.show()
```

&= -\frac{l^{2}}{\sigma^2} \exp\left(-\frac{1}{2} \|\mathbf{Z}_{m} -  \mathbf{Z}_{n}\|^{2} \right)\left[2\left( \mathbf{Z}_{m,i} - \mathbf{Z}_{n,j} \right)\right]

```python
Z = Z / lengthscale
Zsquare = np.sum(np.square(Z), 1)
K_dist2 = -2.*gutil.linalg.tdot(Z) + (Zsquare[:, None] + Zsquare[None, :])
K_dvar = np.exp(-0.5 * K_dist2) # exp(-0.5*||zm - zn||^2 )

lengthscale2 = np.square(lengthscale)

K_dist = 2*(Z[:, None, :] - Z[None, :, :])
dK_dZ = (-variance / lengthscale2) * np.transpose(K_dvar[:, :, np.newaxis] * K_dist, (1, 0, 2))
```
