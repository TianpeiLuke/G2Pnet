```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import scipy as scp
import seaborn as sns
%matplotlib inline 

plt.close('all')
```

```python
size = 160
n = 5
m = 5
G= nx.grid_2d_graph(m,n, periodic=False)  #2d grid graph of mxn nodes, each connected to its nearest neighbors. 


fig= plt.figure(1)
fig.set_size_inches(12.0, 12.0)


ax1 = plt.subplot(321)
pos = nx.spring_layout(G)
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = '''
        Spring layout 
      '''
plt.text(-1,-0.1,txt, fontsize='x-large')
fig.subplots_adjust(wspace=0)

plt.subplot(322, sharey = ax1)
pos = nx.spectral_layout(G) #nx.spring_layout(G)
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = "Spectral layout (" + "n=" + str(n) + ", m=" + str(m)+ ")"

plt.text(0.9,-0.1,txt, fontsize='x-large')
plt.subplots_adjust(hspace = .001)

n = 5
m = 8
G= nx.grid_2d_graph(m,n, periodic=False) 

ax2 = plt.subplot(323)
pos = nx.spring_layout(G)
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = '''
        Spring layout 
      '''
plt.text(-2,-0.1,txt, fontsize='x-large')
fig.subplots_adjust(wspace=1)

plt.subplot(324, sharey = ax2)
pos = nx.spectral_layout(G) #nx.spring_layout(G)
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = "Spectral layout (" + "n=" + str(n) + ", m=" + str(m) + ")"

plt.text(1.5,-0,txt, fontsize='x-large')
plt.subplots_adjust(hspace = .001)


n = 5
m = 1 
G= nx.grid_2d_graph(m,n, periodic=False) 

ax3 = plt.subplot(325)
pos = nx.spring_layout(G)
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = '''
        Spring layout 
      '''
plt.text(-1,-0.1,txt, fontsize='x-large')
fig.subplots_adjust(wspace=1)

plt.subplot(326, sharey = ax3)
pos = nx.spectral_layout(G) #Position nodes using the eigenvectors of the graph Laplacian
ax=nx.draw(G,pos=pos, node_size = size)  
limits=plt.axis('off') # turn of axis
txt = "Spectral layout (" + "n=" + str(n) + ", m=" + str(m) + ")"
 
plt.text(1,-0,txt, fontsize='x-large')

plt.show()
```
